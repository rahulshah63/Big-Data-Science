```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import xgboost as xg
import pandas as pd
```

```python
PublicIP = '54.237.56.69'#@param {type:"string"}
ModalLocation = '/content/drive/My Drive/DataScienceProject/model.h5'#@param {type:"string"}
```

```python
consumer = KafkaConsumer(
    'Sales',
     bootstrap_servers=[PublicIP], 
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
#Modal Loading
xgb = xg.XGBRegressor()
xgb.load_model(ModalLocation)
```

```python
for count, data in enumerate(consumer):
    data_dict = data.value
    df = pd.DataFrame([data_dict])
    # print(data_dict)
    # Dump Data in json file which can be saved to S3
    with open("black_friday_sale_{}.json".format(count), 'w') as file:
        json.dump(data_dict, file) 

    # Make Prediction
    purchase_pred = xgb.predict(df)
    print('Purchase Prediction : $ ', purchase_pred[0])
```
