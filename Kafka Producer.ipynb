```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
PublicIP = '54.237.56.69'#@param {type:"string"}
DataLocation = '/content/drive/My Drive/DataScienceProject/X_test.csv'#@param {type:"string"}
```

```python
producer = KafkaProducer(bootstrap_servers=[PublicIP],
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
# To test
producer.send('Sales', value={'sale':'$10,000'})   #we have topic named as Sales
```

```python
df = pd.read_csv(DataLocation)
```

```python
df.head()
```

```python
while True:
    data = df.sample(1).to_dict(orient="records")[0]  #Sending 5 rows at a time
    producer.send('Sales', value=data)
    sleep(1)
```

```python
producer.flush() #clear data from kafka server
```
