```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
PublicIP = '3.238.16.255:9092'#@param {type:"string"}
DataLocation = '/content/drive/My Drive/DataScienceProject/BlackFriday.csv'#@param {type:"string"}
```

```python
producer = KafkaProducer(bootstrap_servers=[PublicIP],
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
producer.send('Sales', value={'sale':'$10,000'})   #we have topic named as Sales
```

```python
df = pd.read_csv(DataLocation)
```

```python
df.head()
```

```python
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('Sales', value=dict_stock)
    sleep(1)
```

```python
producer.flush() #clear data from kafka server
```
